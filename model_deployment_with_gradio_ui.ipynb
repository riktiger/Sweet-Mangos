```python
import gradio as gr
import keras
import numpy as np
import requests
import PIL

from keras.models import load_model
model=load_model("/content/drive/MyDrive/Colab Notebooks/Projects for Show/Mango Leaf Disease/mango_leaf_disease_detector.h5")
Label=['Anthracnose','Bacterial Canker','Cutting Weevil','Die Back','Gall Midge','Powdery Mildew','Healthy ','Sooty Mould']

def input_img(image):
  rescaled_image= image/255
  resized_image = rescaled_image.reshape(-1,299,299,3)
  detector = model.predict(resized_image)[0]
  return {Label[i]: float(detector[i])for i in range(8)}


Mango_Leaf_Disease_Detector=gr.Interface(fn=input_img,
             inputs=gr.Image(shape=(299,299)),
             outputs=gr.Label(num_top_classes=4))
Mango_Leaf_Disease_Detector.launch(share=True,debug=True)
```
